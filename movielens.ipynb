```
rating_mat = pd.pivot_table(df1 , index ='movieId' , columns ='userId' , values='rating' ).reset_index(drop = True)
```

```
rating_mat.fillna(0 , inplace=True)
```

```
movie_sim = 1 - pairwise_distances(rating_mat.values , metric='correlation')
```

```
movie_sim_df = pd.DataFrame(movie_sim)
```

```
movie_sim_df
```

```
movieids= df[df.movieId == 858]
```

```
movieids
```

```
def similar_movies(movieids , topN  =5):
  moiveidx = df[df.movieId ==movieids].index[0]
  df['similarity'] = movie_sim_df.loc[moiveidx]
  top_n=df.sort_values(by =['similarity'],ascending= False)[0:topN]
  return top_n
```
